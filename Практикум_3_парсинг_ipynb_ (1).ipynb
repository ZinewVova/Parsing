```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
```

```python
def parse_it(url):
  results = dict()
  results['title'] = []
  results['mean_rating'] = []

  for i in range(10):
      url_use = url + f'?page={i}'
      r = requests.get(url_use)
      sleep(3)
      soup = BeautifulSoup(r.text, 'lxml')
      
      elements = soup.find('div', class_ = 'view-content')
      titles = [i.text for i in elements.findAll('div', class_ = 'title')]
      all_rating = [i for i in elements.findAll('div', class_ = 'rating')]
      mean_rating = [i.findAll('span', class_ = 'average-rating')[0].text.replace('Среднее: ','') for i in all_rating]
      
      results['title'].extend(titles)
      results['mean_rating'].extend(mean_rating)
  return pd.DataFrame(results, index = range(1,201))
```

#Фильмы

```python
url = 'https://irecommend.ru/catalog/list/47'
films = parse_it(url)
```

```python
films
```

##сохраняем полученную таблицу в ксв файл

```python
films.to_csv (r'film_reviews.csv', index= False )
```
